What is Pandas?

Pandas is one of the most popular Python libraries for Data Analysis.

It helps you:

✅ Store data in tables
✅ Clean data
✅ Analyze datasets
✅ Work with Excel and CSV files

Import Pandas:

import pandas as pd
Series

A Series is a one-dimensional data structure.

import pandas as pd

s = pd.Series([10, 20, 30, 40])

print(s)

Output:

0    10
1    20
2    30
3    40
dtype: int64
DataFrame

A DataFrame is a two-dimensional table (rows and columns).

import pandas as pd

data = {
    "Name": ["Dhruv", "Aman", "Riya"],
    "Age": [20, 21, 19]
}

df = pd.DataFrame(data)

print(df)

Output:

    Name  Age
0  Dhruv   20
1   Aman   21
2   Riya   19
Rows & Columns
print(df.shape)

Output:

(3, 2)

Meaning:

3 Rows
2 Columns
Shape

Returns total rows and columns.

print(df.shape)
Columns

Returns all column names.

print(df.columns)

Output:

Index(['Name', 'Age'], dtype='object')
Info()

Provides dataset information.

print(df.info())

Shows:

Number of rows
Column names
Data types
Memory usage
Describe()

Provides statistical summary.

print(df.describe())

Output includes:

Count
Mean
Min
Max
Standard Deviation

```python
# Create a DataFrame with 5 students and their marks.
d = {"Name" : ["Dhruv" , "Aarush" , "Riya" , "Priya" , "Komal"] ,
     "Marks" : [ 94, 90 , 89 , 78 , 90]}
df = pd.DataFrame(d)
df
```

```python
# Find the number of rows and columns.
df.shape
```

```python
# Create a DataFrame for employee data.
employee = {"Name" : ["Dhruv" , "Aarush" , "Riya" , "Priya" , "Komal"] ,
     "Department" : [ "IT", "HR" , "IT" , "Sales" , "IT"],
           "Salary" : [60000 , 30000, 40000, 50000 , 45000]}

df = pd.DataFrame(employee)
df
```

```python
# Print only employee names.
df["Name"]
```

```python
# Find the average salary using Pandas.
df["Salary"].mean()
```

```python
# Display statistical summary using describe().
df.describe()
```

```python
# Create a DataFrame with at least 10 rows and explore shape, columns, info(), and describe().

employee = {"Name" : ["Dhruv" , "Aarush" , "Riya" , "Priya" , "Komal","Rahul" , "Neha" , "Sharman" , "Rishabh" , "Aniket"] ,
     "Department" : [ "IT", "HR" , "IT" , "Sales" , "IT" , "IT", "HR" , "IT" , "Sales" , "IT"],
           "Salary" : [60000 , 30000, 40000, 50000 , 45000 , 60000 , 30000, 40000, 50000 , 45000]}

df = pd.DataFrame(employee)
print(df.shape)
print(df.columns)
print(df.info())
print(df.describe())
```

```python
# Student Performance Analyzer

# Features:

# ✅ Create a DataFrame of students
# ✅ Display all data
# ✅ Show average marks
# ✅ Show highest marks
# ✅ Show lowest marks
# ✅ Print DataFrame shape

students = {
    "Name": ["Dhruv", "Aman", "Riya", "Priya", "Karan"],
    "Marks": [85, 90, 88, 92, 80]
}

df = pd.DataFrame(students)
print(df)
print("Avg : ",df["Marks"].mean())
print("Hightest  :",df["Marks"].max())
print("Lowest : ",df["Marks"].min())
print(df.shape)
```
