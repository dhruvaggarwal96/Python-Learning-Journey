Day 16 – Reading CSV Files with Pandas
1️⃣ Reading CSV Files

CSV (Comma Separated Values) files are one of the most common data formats used in Data Analysis.

Pandas provides the read_csv() function to load CSV files into a DataFrame.

import pandas as pd

df = pd.read_csv("sales.csv")
print(df)
2️⃣ View First Rows

Use head() to see the first few records.

print(df.head())

Output:

   OrderID Product  Sales
0        1   Mouse    500
1        2 Keyboard 1200
...
3️⃣ View Last Rows

Use tail() to see the last few records.

print(df.tail())
4️⃣ View Random Rows

Use sample() to inspect random records.

print(df.sample(5))
5️⃣ Check Dataset Shape

Returns:

Number of rows
Number of columns
print(df.shape)

Output:

(1000, 5)

Meaning:

1000 Rows
5 Columns
6️⃣ Get Dataset Information

Use info() to check:

Column names
Data types
Non-null values
Memory usage
df.info()
7️⃣ Check Missing Values

Missing values are empty cells in the dataset.

print(df.isnull().sum())

Example Output:

Product     0
Sales       2
Profit      1
dtype: int64

```python
import pandas as pd

df = pd.read_csv("amazon.csv")

print("Rows and Columns:")
print(df.shape)

print("\nColumn Names:")
print(df.columns)

print("\nData Types:")
print(df.dtypes)

print("\nMissing Values:")
print(df.isnull().sum())

print("\nFirst 10 Rows:")
print(df.head(10))
```

```python
# Load a CSV file named amazon.csv

import pandas as pd

df = pd.read_csv("amazon.csv")
```

```python
# Print the first 10 rows.
df.head(10)
```

```python
# Print the last 3 rows.
df.tail(3)
```

```python
# Display 5 random rows.
df.sample(5)
```

```python
# Find the number of rows and columns.
row = df.shape[0]
col = df.shape[1]
print("Number of Rows : " , row)
print("Number of Columns : " , col)
```

```python
# Print all column names.
df.columns
```

```python
# Check the data type of each column.
df.info()
```

```python
# Count missing values in each column.
df.isnull().sum()
```

```python
# Find columns that have missing values.

missing_column = df.columns[df.isnull().sum() > 0]
print("Columns With the Missing Values :")
print(missing_column)
```

```python
# Display information about the dataset.
df.info()
```
