```python
What is NumPy?

NumPy (Numerical Python) is a Python library used for:

Mathematical calculations
Data analysis
Machine Learning
Working with large datasets
Why NumPy?

Python lists are slower for numerical operations.

marks = [78, 85, 90, 67, 88]

NumPy arrays are optimized for speed:

import numpy as np

marks = np.array([78, 85, 90, 67, 88])

Benefits:

✅ Faster

✅ Less memory

✅ Easy calculations

🔑 Important Concepts
Create Array
import numpy as np

arr = np.array([10, 20, 30, 40, 50])

print(arr)

Output:

[10 20 30 40 50]
Array Operations
arr = np.array([1, 2, 3, 4, 5])

print(arr + 10)
print(arr * 2)
print(arr / 2)

Output:

[11 12 13 14 15]

[ 2  4  6  8 10]

[0.5 1.  1.5 2.  2.5]
Useful Functions
arr = np.array([78, 85, 90, 67, 88])

print(np.mean(arr))
print(np.max(arr))
print(np.min(arr))
print(np.sum(arr))
Indexing
arr = np.array([10, 20, 30, 40, 50])

print(arr[0])
print(arr[-1])

Output:

10
50
Slicing
arr = np.array([10, 20, 30, 40, 50])

print(arr[1:4])

Output:

[20 30 40]
```
