```python
What is OOP?

OOP (Object-Oriented Programming) is a programming paradigm that organizes code using classes and objects.

Real-Life Example
Class = Car (Blueprint)
Object = BMW, Audi, Tesla (Actual cars created from the blueprint)

Think of a class as a design/template and an object as a real thing created from that design.

🏗️ Class Syntax
class Student:
    pass

pass means the class is empty for now.

🎯 Creating an Object
class Student:
    pass

s1 = Student()

print(type(s1))

Output:

<class '__main__.Student'>

🎤 Interview Questions & Answers
1. What is OOP?

OOP (Object-Oriented Programming) is a programming approach that organizes code using classes and objects.

2. What is a Class?

A class is a blueprint or template used to create objects.

Example:

class Student:
    pass
3. What is an Object?

An object is an instance of a class.

Example:

s1 = Student()

Here s1 is an object of the Student class.

4. Why use OOP?
Better code organization
Reusability
Easier maintenance
Real-world modeling
Scalability for large projects
```
