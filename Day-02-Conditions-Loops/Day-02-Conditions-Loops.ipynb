## Python : Conditions & Loops in Python

1️⃣ Conditional Statements (if, elif, else)

Conditional statements help a program make decisions.

Real-Life Example

Imagine you are checking your exam marks:

If marks ≥ 90 → Grade A
If marks ≥ 75 → Grade B
Otherwise → Grade C

Python uses:

if condition:
    # code

elif condition:
    # code

else:
    # code
if Statement

Executes code only when the condition is True.

age = 18

if age >= 18:
    print("You can vote")
Output
You can vote
How it works
Python checks age >= 18
Result = True
Code inside if executes
if-else Statement

When there are only two possibilities.

age = 15

if age >= 18:
    print("You can vote")
else:
    print("You cannot vote")
Output
You cannot vote
if-elif-else Statement

Used when there are multiple conditions.

marks = 82

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
else:
    print("Grade C")
Output
Grade B
Flow
marks = 82

82 >= 90 ? No

82 >= 75 ? Yes

Print Grade B
Comparison Operators
Operator	Meaning
==	Equal to
!=	Not equal to
>	Greater than
<	Less than
>=	Greater than or equal to
<=	Less than or equal to

Example:

x = 10

print(x == 10)
print(x != 5)
print(x > 20)

Output:

True
True
False
Logical Operators
and

Both conditions must be True.

age = 20
has_id = True

if age >= 18 and has_id:
    print("Entry Allowed")
or

At least one condition must be True.

age = 15
special_permission = True

if age >= 18 or special_permission:
    print("Entry Allowed")
not

Reverses the result.

is_raining = False

if not is_raining:
    print("Go outside")

4️⃣ while Loop

Runs as long as the condition remains True.

Syntax
while condition:
    code
Example
count = 1

while count <= 5:
    print(count)
    count += 1

Output:

1
2
3
4
5
How it Works
count = 1

1 <= 5 → print 1

count = 2

2 <= 5 → print 2

...

count = 6

6 <= 5 → False

Loop stops
Infinite Loop

Be careful:

while True:
    print("Hello")

This runs forever because the condition never becomes False.

5️⃣ break

Used to stop a loop immediately.

Example:

for i in range(10):

    if i == 5:
        break

    print(i)

Output:

0
1
2
3
4
What happened?

When i becomes 5:

break

The loop ends instantly.

6️⃣ continue

Skips the current iteration and moves to the next one.

Example:

for i in range(6):

    if i == 3:
        continue

    print(i)

Output:

0
1
2
4
5
What happened?

When i becomes 3:

continue

Python skips printing 3 and moves to the next iteration.

Difference Between break and continue
break	continue
Stops the loop completely	Skips only current iteration
Loop ends	Loop continues
Used when work is finished	Used when one iteration should be ignored

Example:

# break
for i in range(5):
    if i == 2:
        break
    print(i)

Output:

0
1
# continue
for i in range(5):
    if i == 2:
        continue
    print(i)

Output:

0
1
3
4

```python
# ⭐ Challenge Questions
# Q23. FizzBuzz (Famous Interview Question)

# For numbers 1 to 20:

# If divisible by 3 → Print "Fizz"
# If divisible by 5 → Print "Buzz"
# If divisible by both → Print "FizzBuzz"
# Otherwise print number

for i in range(1,21):
    if i%3 == 0:
        print("Fizz")
    elif i%5 == 0:
        print("Buzz")
    elif (i%5 == 0) and (i%3 == 0):
        print("FizzBuzz")
    else:
        print(i)
```
