```python
import numpy as np 

students = np.array(["Aman" , 
                     "Rahul",
                     "Aarush",
                     "Ujjwal",
                     "Divya",
                     "Riya",])

marks = np.array([65,
                 75,
                 78,
                 89,
                 54,
                 98])
```

```python
for i in range(len(students)):
    print(students[i] , ":" , marks[i])
```

```python
print("Average Marks : " , np.mean(marks))
print("Highest Marks : " , np.max(marks))
print("Lowest Marks : " , np.min(marks))
print("Total Marks : " , np.sum(marks))
```

```python
topper_index = np.argmax(marks)
print("Topper :" , students[topper_index]) 
print("Marks :" , marks[topper_index])
```

```python
average = np.mean(marks)
print("Students Above Average")
for i in range(len(students)):
    if marks[i] > average:
        print(students[i])
```

```python
print("Students Grades :")
for i in range(len(students)):
    if 90<=marks[i]<=100 :
        print(f"{students[i]} : {marks[i]} -> A+")
    elif 80<=marks[i]<=89 :
        print(f"{students[i]} : {marks[i]} -> A")
    elif 70<=marks[i]<=79 :
        print(f"{students[i]} : {marks[i]} -> B")
    elif 60<=marks[i]<=69 :
        print(f"{students[i]} : {marks[i]} -> C")
    else :
        print(f"{students[i]} : {marks[i]} -> D")
```

```python
print("Count Students in Each Grade")
grades = {
    "a_plus": 0,
    "a": 0,
    "b": 0,
    "c": 0,
    "d": 0
}
for mark in marks:
    if mark >=90:
        grades["a_plus"] += 1
    elif mark >=80:
        grades["a"] += 1
    elif mark >=70:
        grades["b"] += 1
    elif mark >= 60:
        grades["c"] += 1
    else:
        grades["d"] += 1
print("Grade Summary " )
print("A+ :", grades["a_plus"])
print("A  :", grades["a"])
print("B  :", grades["b"])
print("C  :", grades["c"])
print("D  :", grades["d"])
```
