## Python Basic - Variables, Data Types, Input/Output, Operators, and Type Casting

Python Basics: Variables, Data Types, Input/Output, Operators, and Type Casting

These are the fundamental building blocks of Python. Understanding them well will make learning Python much easier.

1. Variables

A variable is a container that stores data in memory.

Think of a variable as a labeled box where you keep information.

Syntax
variable_name = value
Example
name = "Dhruv"
age = 20
salary = 50000

Here:

name stores "Dhruv"
age stores 20
salary stores 50000
How Variables Work
x = 10
print(x)

Output:

10

You can change the value later:

x = 10
x = 20

print(x)

Output:

20

The old value is replaced.

Variable Naming Rules
Valid
student_name = "Rahul"
age1 = 20
_marks = 95
Invalid
1age = 20      # Cannot start with number
student-name   # Hyphen not allowed
class = 10     # Keyword not allowed
Multiple Variables
a = 10
b = 20
c = 30

or

a, b, c = 10, 20, 30
2. Data Types

A data type tells Python what kind of data is stored in a variable.

Main Data Types
Data Type	Description	Example
int	Integer	10
float	Decimal Number	10.5
str	Text/String	"Hello"
bool	True/False	True
list	Collection of items	[1,2,3]
tuple	Immutable collection	(1,2,3)
set	Unique items	{1,2,3}
dict	Key-value pairs	{"name":"Dhruv"}
Integer (int)

Whole numbers.

age = 20
marks = 95
print(type(age))

Output:

<class 'int'>
Float

Numbers with decimal points.

height = 5.9
price = 99.99
print(type(height))

Output:

<class 'float'>
String (str)

Text enclosed in quotes.

name = "Dhruv"
city = 'Delhi'
print(type(name))

Output:

<class 'str'>
Boolean (bool)

Only two values:

True
False

Example:

is_student = True
print(type(is_student))

Output:

<class 'bool'>
List

Stores multiple values.

fruits = ["Apple", "Banana", "Mango"]

Access item:

print(fruits[0])

Output:

Apple
Tuple

Similar to list but cannot be changed.

coordinates = (10, 20)
Set

Stores unique values only.

numbers = {1, 2, 3, 4}

Duplicate values are removed automatically.

s = {1,1,1,2,2,3}
print(s)

Output:

{1,2,3}
Dictionary

Stores data in key-value format.

student = {
    "name": "Dhruv",
    "age": 20,
    "marks": 90
}

Access value:

print(student["name"])

Output:

Dhruv
3. Input and Output

Input means taking data from the user.

Output means displaying data to the user.

Output Using print()
print("Hello World")

Output:

Hello World
Printing Variables
name = "Dhruv"

print(name)

Output:

Dhruv
Multiple Values
name = "Dhruv"
age = 20

print(name, age)

Output:

Dhruv 20
Taking Input
name = input("Enter your name: ")

If user enters:

Dhruv

Then:

name = "Dhruv"
Example
name = input("Enter Name: ")

print("Welcome", name)

Input:

Dhruv

Output:

Welcome Dhruv
Important Note

input() always returns a string.

Example:

age = input("Enter Age: ")

print(type(age))

Output:

<class 'str'>

Even if user enters:

20

it is treated as "20".

4. Operators

Operators perform operations on values.

Arithmetic Operators
Operator	Meaning
+	Addition
-	Subtraction
*	Multiplication
/	Division
//	Floor Division
%	Modulus
**	Power
Example
a = 10
b = 3

print(a + b)
print(a - b)
print(a * b)
print(a / b)
print(a // b)
print(a % b)
print(a ** b)

Output:

13
7
30
3.3333
3
1
1000
Comparison Operators

Used for comparisons.

Operator	Meaning
==	Equal
!=	Not Equal
>	Greater Than
<	Less Than
>=	Greater or Equal
<=	Less or Equal

Example:

a = 10
b = 20

print(a == b)
print(a < b)

Output:

False
True
Logical Operators
Operator	Meaning
and	Both True
or	At Least One True
not	Reverse Result

Example:

a = True
b = False

print(a and b)
print(a or b)
print(not a)

Output:

False
True
False
Assignment Operators
x = 10

Shortcut operators:

x += 5

Same as:

x = x + 5

Other examples:

x -= 5
x *= 5
x /= 5
5. Type Casting

Type casting means converting one data type into another.

Why Type Casting?

Remember:

input()

returns a string.

Example:

a = input("Enter first number: ")
b = input("Enter second number: ")

print(a + b)

Input:

10
20

Output:

1020

Because Python joins strings.

Converting String to Integer
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

print(a + b)

Input:

10
20

Output:

30
Common Type Casting Functions
Function	Converts To
int()	Integer
float()	Float
str()	String
bool()	Boolean
String to Integer
num = "100"

num = int(num)

print(type(num))

Output:

<class 'int'>
Integer to Float
x = 10

y = float(x)

print(y)

Output:

10.0
Float to Integer
x = 10.9

y = int(x)

print(y)

Output:

10

Decimal part is removed.

Integer to String
age = 20

age = str(age)

print(type(age))

Output:

<class 'str'>
Type Casting Example
name = input("Enter Name: ")
age = int(input("Enter Age: "))
salary = float(input("Enter Salary: "))

print("Name:", name)
print("Age:", age)
print("Salary:", salary)

Input:

Dhruv
20
50000.5

Output:

Name: Dhruv
Age: 20
Salary: 50000.5
Complete Example Program
name = input("Enter your name: ")
age = int(input("Enter your age: "))
marks = float(input("Enter your marks: "))

print("\nStudent Details")
print("Name :", name)
print("Age :", age)
print("Marks :", marks)

if marks >= 40:
    print("Result : Pass")
else:
    print("Result : Fail")
Sample Output
Enter your name: Dhruv
Enter your age: 20
Enter your marks: 85.5

Student Details
Name : Dhruv
Age : 20
Marks : 85.5
Result : Pass
Quick Summary
Concept	Purpose

Input	Takes data from user

Operators	Perform operations
Type Casting	Converts data from one type to another

A simple flow in Python is:

User Input
     ↓
Type Casting
     ↓
Variables Store Data
     ↓

     ↓
print() Displays Output
