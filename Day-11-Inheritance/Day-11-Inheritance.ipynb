```python
🚀 Day 11 — Inheritance in Python
📚 Study Notes
What is Inheritance?

Inheritance allows one class to acquire the properties and methods of another class.

It helps in:

Code Reusability
Reducing Redundant Code
Better Organization of Programs
Parent Class (Base Class)
class Animal:
    def sound(self):
        print("Animal makes a sound")
Child Class (Derived Class)
class Dog(Animal):
    pass

dog = Dog()
dog.sound()

Output:

Animal makes a sound

The Dog class inherited the sound() method from Animal.

Reusing Attributes
class Animal:
    def __init__(self, name):
        self.name = name

class Dog(Animal):
    pass

dog = Dog("Tommy")
print(dog.name)

Output:

Tommy
Method Overriding

A child class can replace a method from the parent class.

class Animal:
    def sound(self):
        print("Animal makes a sound")

class Dog(Animal):
    def sound(self):
        print("Dog barks")

dog = Dog()
dog.sound()

Output:

Dog barks
Using super()

super() allows the child class to access parent class methods.

class Animal:
    def __init__(self, name):
        self.name = name

class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name)
        self.breed = breed

dog = Dog("Tommy", "Labrador")

print(dog.name)
print(dog.breed)
```

```python
# Exercise 1: Vehicle Inheritance

# Create:

# class Vehicle

# Attribute:

# brand

# Method:

# show_brand()

# Create child class:

# class Car

# Method:

# show_car()

# Create object and test both methods.

class Vehicle:
    def __init__(self , brand):
        self.brand = brand
    def show_brand(self):
        print("The Brand is : " , self.brand)

class Car(Vehicle):
    def __init_(self):
        pass
    def show_car(self):
        print("There is a car")
v1 = Car("suzuki")

v1.show_car()
v1.show_brand()
```

```python
# Exercise 2: Employee System

# Create:

# class Employee

# Attributes:

# name
# salary

# Method:

# display()

# Create child class:

# class Manager

# Additional Attribute:

# department

# Display all details.

class Employee:
    def __init__(self , name , salary ) :
        self.name = name
        self.salary = salary
    def display(self):
        print("Employee Name:" , self.name)
        print("Employee Salary:" , self.salary)
class Manager(Employee):
    def department(self , department):
        self.department = department
        print("Employee Department : " , self.department)
e = Manager("dhruv" , 30000)
e.display()
e.department("IT")
```

```python
# Exercise 3: Shape Calculator

# Create:

# class Shape

# Method:

# area()

# Create:

# class Rectangle

# Override area() and calculate:

# length × width

class Shape:
    def area(self):
        return 0
class Rectangle(Shape):
    def __init__(self , length , breadth ):
        self.length = length 
        self.breadth = breadth 
    def area(self):
        return self.length * self.breadth
k = Rectangle(10,10)
print(k.area())
```

```python
# Exercise 4: Student Management

# Create:

# class Person

# Attributes:

# name
# age

# Create:

# class Student

# Additional Attribute:

# course

# Display all information.
class Person:
    def __init__(self , name , age):
        self.name = name
        self.age = age
class Student(Person):
    def course(self , course):
        self.course = course
    def display_information(self):
        print(f"Student Name : {self.name} \n Student Age : {self.age} \n Student Course : {self.course}")
p = Student("Dhruv" , 19 )
p.course("Btech")
p.display_information()
```

```python
# 🚀 Mini Project — School Management System
# Requirements

# Create Parent Class:

# class Person

# Attributes:

# name
# age

# Method:

# display_info()
# Child Class 1
# class Student(Person)

# Additional Attribute:

# grade

# Method:

# show_student()
# Child Class 2
# class Teacher(Person)

# Additional Attribute:

# subject

# Method:

# show_teacher()
# Example Output
# Student Details
# Name: Dhruv
# Age: 19
# Grade: A

# Teacher Details
# Name: Sharma
# Age: 35
# Subject: Mathematics

class Person:
    def __init__(self , age , name ) : 
        self.name  = name 
        self.age = age
    def display_info(self):
        print(f"Name : {self.name} \nAge : {self.age}")
class Student(Person):
    def __init__(self ,age , name ,  grade):
        super().__init__(age , name) 
        self.grade = grade
    def show_student(self):
        print("Student Details")
        print(f"Name : {self.name} \nAge : {self.age} \nGrade : {self.grade}")
class Teacher(Person):
    def __init__(self , subject , age , name  ) :
        super().__init__(age , name)
        self.subject = subject 
    def show_teacher(self):
        print("Teacher Details")
        print(f"Name : {self.name} \nAge : {self.age} \nSubject : {self.subject}")
s = Student(19 , "Dhruv" , "A" )
s.show_student()
print()
k = Teacher("English" , 19 , "Dhruv")
k.show_teacher()
```
