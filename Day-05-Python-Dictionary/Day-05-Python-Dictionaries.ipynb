## Day 5 Python Dictionaries

1️⃣ Creating a Dictionary
student = {
    "name": "Dhruv",
    "age": 21,
    "city": "Delhi"
}

print(student)

Output:

{'name': 'Dhruv', 'age': 21, 'city': 'Delhi'}

2️⃣ Accessing Values Using Keys
print(student["name"])
print(student["age"])

Output:

Dhruv
21

3️⃣ Adding New Data
student["course"] = "Data Science"

print(student)

Output:

{
 'name': 'Dhruv',
 'age': 21,
 'city': 'Delhi',
 'course': 'Data Science'
}
4️⃣ Updating Existing Data

student["age"] = 22

print(student)

Output:

22
5️⃣ Using keys()
print(student.keys())

Output:

dict_keys(['name', 'age', 'city', 'course'])
6️⃣ Using values()
print(student.values())

Output:

dict_values(['Dhruv', 22, 'Delhi', 'Data Science'])
7️⃣ Using items()
print(student.items())

Output:

dict_items([
 ('name', 'Dhruv'),
 ('age', 22),
 ('city', 'Delhi'),
 ('course', 'Data Science')
])

```python
# Q12. Create a dictionary of monthly sales:

sales = {
    "Jan": 5000,
    "Feb": 7000,
    "Mar": 6500
}

# Find:

# Total Sales
# Highest Sales Month
# Lowest Sales Month
total = 0
highest = 0
lowest = 9999999999999999999999999999
for i in sales:
    total = total + sales[i]
    if highest<sales[i]:
        highest = sales[i]
    if lowest>sales[i]:
        lowest = sales[i]
print("Total sales :" , total)
print("Lowest sales : " , lowest)
print("Hightest sales : " ,highest)
```

```python
# Q13. Create a frequency counter for:

fruits = ["apple", "banana", "apple", "orange", "banana", "apple"]

# Expected Output:

# {
#  'apple': 3,
#  'banana': 2,
#  'orange': 1
# }
count = 0 
for i in fruits:
    k = fruits.count(i)
    if k > 1:
        for j in range(0,k):
            fruits.remove(i)
    print(f"{i} : {k}")
```

```python
# 🚀 Mini Project: Student Report Card

# Create a program that:

# Stores student details in a dictionary.
# Stores marks for 5 subjects.
# Calculates:
# Total Marks
# Average Marks
# Highest Subject
# Lowest Subject
# Prints a report card.

# Example Output:

# ===== REPORT CARD =====

# Name: Dhruv

# Math: 90
# Science: 85
# English: 88
# Python: 95
# SQL: 92

# Total: 450
# Average: 90.0

# Highest Subject: Python
# Lowest Subject: Science
print("=====Student Report Card====\n\n")
n = int(input("Number Of Records Want to Enter : " ))
d = {}    
for i in range(1,n + 1):
    name = input(f"Enter the {i} Name Of Student : ")
    marks = []
    for k in range(1,6):
        mark = int(input(f"Enter The marks of {k} Subject : " ))
        marks.append(mark)
    d[name] = marks
for name in d:
    print("\n===== Report Card ====\n")
    print(f"Name : {name}\n")
    total = 0
    for i in range(0,5):
        total = total + d[name][i]
    print(f"Total Marks : {total}")
    print(f"Avg Marks : {total/5}")
    print(f"Highest Marks  : {max(d[name])}")
    print(f"Lowest Marks  : {min(d[name])}")
```
