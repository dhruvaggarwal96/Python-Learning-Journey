📚 Topics Covered
1️⃣ What is a Function?

A function is a block of reusable code that performs a specific task.

def greet():
    print("Hello, World!")

greet()
2️⃣ Function with Parameters

Parameters allow us to pass data into a function.

def greet(name):
    print(f"Hello, {name}")

greet("Dhruv")

Output:

Hello, Dhruv
3️⃣ Function with Return Statement

The return keyword sends a value back to the caller.

def add(a, b):
    return a + b

result = add(5, 3)
print(result)

Output:

8
4️⃣ Local Variables

Variables created inside a function are local to that function.

def test():
    x = 10
    print(x)

test()
5️⃣ Global Variables

Variables created outside a function can be accessed globally.

name = "Dhruv"

def show_name():
    print(name)

show_name()
6️⃣ Small Project: Calculator Function
def calculator(a, b):
    print("Addition:", a + b)
    print("Subtraction:", a - b)
    print("Multiplication:", a * b)
    print("Division:", a / b)

calculator(10, 5)
💡 Key Takeaways
Functions make code reusable.
Parameters allow input to functions.
Return statements provide output from functions.
Local variables exist only inside functions.
Global variables can be accessed throughout the program.
Functions are essential for Data Science and Machine Learning projects.

```python
# Q12. Grade Calculator

# Create a function that accepts marks and returns:

# Marks	Grade
# 90+	      A
# 80-89	  B
# 70-79	  C
# Below 70  D

marks = float(input("Enter the Marks : " ))
def grade(marks):
    if marks >= 90:
        return "A"
    elif 80<=marks<=89:
        return "B"
    elif 70<=marks<=79:
        return "C"
    else:
        return "D"
a = grade(marks)
print("Grade is ",a)
```

```python
# Q13. Leap Year Checker

# Create a function that accepts a year and returns whether it is a leap year.
year = int(input("Enter the Year:"))
def is_leap(year):
    if (year%4 == 0 and year%100 !=0) or (year%400 == 0):
        return "Leap Year"
    else:
        return "Not Leap Year"
k = is_leap(year)
print(k)
```

```python
# Q15. Largest Among Three Numbers

# Create a function using conditions.
a = int(input("Enter the Number : "))
b = int(input("Enter the Number : "))
c = int(input("Enter the Number : "))
def largest(a,b,c):
    if a>b and a>c:
        print(f"{a} is largest")
    elif b>a and b>c:
        print(f"{b} is largest")
    else:
        print(f"{c} is largest")
largest(a,b,c)
```

```python
# Q16. Sum of Numbers 1 to N
n = int(input("Enter the Number : "))
def sum_n(n):
    k = 0 
    for i in range(1,n+1):
         k = k + n
    return k
z = sum_n(n)
print(z)
```

```python
# Q18. Count Vowels

# Create a function that counts vowels in a string.
a = input("Enter The String : " )
k = ['a','e','i','o','u']
count = 0
def vowel_count(k,a,count):
    for i in a:
        # print(type(i))
        if i in k:
            count = count + 1
    return count
l = vowel_count(k,a,count)
print(l)
```

```python
# Q19. Factorial Function
n = int(input("Enter the Number : " ))
def fact(n):
    k = 1
    for i in range(1,n+1):
        k = k*i
    print(k)
fact(n)
```

```python
# Q20. Prime Number Checker

# Create a function that returns True or False.
n = int(input("Enter The Number : " ))
def is_prime(n):
    is_p = True
    for i in range(2,n):
        if n%i == 0:
            print("Number is not Prime")
            is_p = False
        if is_p == False:
            break
    else:
        print("number is prime")
is_prime(n)
```
