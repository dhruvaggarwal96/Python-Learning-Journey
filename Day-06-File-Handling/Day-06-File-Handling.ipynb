📅 Day 6 — File Handling in Python
🎯 Why File Handling?

Until now, your data disappears when the program stops.

Example:

name = "Dhruv"

When the program closes, the value is gone.

Files allow us to store data permanently.

Examples:

User records
Student data
Login details
Project data
Data Science datasets (CSV files)
1️⃣ Creating a File
file = open("myfile.txt", "w")
file.close()
Explanation
"w" = Write mode
Creates file if it doesn't exist
Deletes old content if file already exists
2️⃣ Writing Data to a File
file = open("myfile.txt", "w")

file.write("Hello Dhruv!")

file.close()

After running:

myfile.txt

contains:

Hello Dhruv!
3️⃣ Reading a File
file = open("myfile.txt", "r")

content = file.read()

print(content)

file.close()

Output:

Hello Dhruv!
4️⃣ Appending Data

Appending means adding data without deleting old data.

file = open("myfile.txt", "a")

file.write("\nWelcome to Python")

file.close()

File becomes:

Hello Dhruv!
Welcome to Python
5️⃣ Using with (Professional Method)

Instead of:

file = open(...)
...
file.close()

Use:

with open("myfile.txt", "r") as file:
    content = file.read()
    print(content)

Benefits:

✅ Cleaner

✅ Automatically closes file

✅ Industry standard

6️⃣ Reading Line by Line

Suppose file contains:

Dhruv
Rahul
Aman

Code:

with open("students.txt", "r") as file:
    for line in file:
        print(line.strip())

Output:

Dhruv
Rahul
Aman
7️⃣ Handling File Not Found Error
try:
    with open("abc.txt", "r") as file:
        print(file.read())

except FileNotFoundError:
    print("File does not exist.")

Output:

File does not exist.

```python
# 🚀 Main Project (Today's Assignment)
# Student Data Storage System
# Requirements:
# Menu
# 1. Add Student
# 2. View Students
# 3. Exit
# Functionality
# Add Student
# Enter Name: Dhruv
# Save in file.
# View Students
# Show all stored students.
# Exit
print("======= STUDENT DATA STORAGE SYSTEM =======")
while True:
  try:
    choice = int(input("Choose The Option : \n    1.Add Student \n    2.View Students \n    3.Exit \nEnter Choice : "))
    if choice == 1:
      name = input("Enter The Name Of the Student : ")
      with open("student.txt","a") as file:
        file.write(name + "\n")
      print("Record Saved Successfully")
    elif choice == 2:
      with open('student.txt',"r") as file:
        content = file.read()
        print(content)
    elif choice == 3:
      break
    else:
      print("Invalid Choice")
  except FileNotFoundError:
      with open("student.txt","w") as file:
        pass
```

```python
# Question 8: Search Student

# Ask the user for a student name and check whether it exists in students.txt.

# Example:

# Enter Name: Rahul

# Output:

# Student Found

# or

# Student Not Found

name = input("Enter the Student Name Want to Search : ")
with open("students.txt","r") as file:
  content = file.read()
  if name in content:
    print("student Found ")
  else:
    print("Student Not found")
```

```python
# Question 10: Store Multiple Names

# Ask the user:

# How many students?

# Then store all names in a file.

# Example:

# How many students? 3

# Enter Name: Dhruv
# Enter Name: Rahul
# Enter Name: Aman
n = int(input("Enter Number of Students Want to Save : "))
with open("students.txt","a") as file:
  for i in range(n):
    name = input("Enter the name of the student :")
    file.write(name + "\n")

```

```python
# Question 11: Marks Storage System

# Store student marks in a file.

# Format:

# Dhruv,89
# Rahul,76
# Aman,92
name = input("Enter the name of the student : ")
marks = int(input("Enter the marks of the student :"))
with open("marks.txt","a") as file:
  file.write(f"{name},{marks}\n")
```

```python
total = 0
with open("marks.txt","r") as file:
  content = file.readlines()
  print(content)
  for number in content:
    total = total + int(number[-3:])
  print("Avg is : ", total/len(content))
```

```python
maxi = 0
with open("marks.txt","r") as file:
  content = file.readlines()
  print(content)
  for number in content:
    if int(number[-3:]) > maxi:
      maxi = int(number[-3:])
  print("Max is : ", maxi)
```
