📖 Session 1 Notes: Understanding Exceptions
What is an Exception?

An Exception is an error that happens while a program is running.

Example:

print(10 / 0)

Output:

ZeroDivisionError

The program stops immediately.

Why Use Exception Handling?

Without exception handling:

age = int(input("Enter age: "))
print(age)

If user enters:

abc

Program crashes:

ValueError

With exception handling:

try:
    age = int(input("Enter age: "))
    print(age)

except ValueError:
    print("Please enter a valid number")

Program handles the error safely.

try Block

Code that might cause an error goes inside try.

try:
    num = 10 / 0
except Block

Runs when an error occurs.

try:
    num = 10 / 0

except:
    print("Something went wrong")

Output:

Something went wrong
Handling Specific Exceptions
ZeroDivisionError
try:
    num = 10 / 0

except ZeroDivisionError:
    print("Cannot divide by zero")
ValueError
try:
    age = int(input("Enter age: "))

except ValueError:
    print("Enter valid number")
📖 Session 2 Notes: else and finally
else

Runs only if no error occurs.

try:
    num = 10 / 2

except ZeroDivisionError:
    print("Error")

else:
    print("Division Successful")

Output:

Division Successful
finally

Runs whether an error occurs or not.

try:
    num = 10 / 0

except:
    print("Error")

finally:
    print("Program Finished")

Output:

Error
Program Finished
Complete Structure
try:
    # risky code

except:
    # handle error

else:
    # if no error

finally:
    # always runs

```python
# Q4

# Create a calculator that performs:

# +
# -
# *
# /

# using exception handling.
while True:
  try:
    a = int(input("Enter the First Number : "))
    b = int(input("Enter the Second Number : "))
    choice = int(input("Choose : \n    1.Addition\n    2.Subtraction\n    3.Division\n    4.Multiplication\n    5.Exit\nEnter Choice :  "))
    if choice == 1:
      print("Addition is :" , a+b)
    elif choice == 2:
      print("Subtraction is :" , a-b)
    elif choice == 3:
      print("Division is :" , a/b)
    elif choice == 4:
      print("Multiplication is :" , a*b)
    elif choice == 5:
      break
    else:
      print("Invalid Choice")
  except ZeroDivisionError:
    print("Cannot Divide by Zero")
  except ValueError:
    print("Please Enter Valid Numbers")
  except:
    print("Something Went Wrong")
```

```python
# Marks Analyzer

# Read marks from a file.

# Handle:

# Missing file
# Invalid values

# Calculate:

# Average
# Highest
# Lowest
# try:
try:
  total = 0
  count = 0
  high = 0
  low = float("inf")

  with open("marks.txt","r") as file:
      content = file.readlines()

      for i in content:
        try:
          mark = int(i.strip())
          total += mark
          count += 1
          if mark > high:
            high = mark
          if mark < low:
            low = mark
        except ValueError:
          print(f"Invalid Value found: {i.strip()}")

except FileNotFoundError:
  print("File Not Found")
else:
  print(f"Average : {total/count}")
  print(f"Highest : {high}")
  print(f"Lowest : {low}")
finally:
  print("Thanks for using the program")
```

```python
# Sales Data Reader

# Read:

# sales.txt

# Handle:

# Missing file
# Incorrect data

# Calculate:

# Total Sales
# Average Sales

try:
  with open("sales.txt","r") as file:
    total = 0
    count = 0
    content = file.readlines()

    for i in content:
      try:
        sale = int(i.strip())
        total += sale
        count += 1
      except ValueError:
        print("Invalid Data")

    if count > 0:
        print(f"Total Sales : {total}")
        print(f"Average Sales : {total/count}")
    else:
      print("No Sales Data")
except FileNotFoundError:
  print("File not found")
finally:
  print("program run succesfully:")
```
