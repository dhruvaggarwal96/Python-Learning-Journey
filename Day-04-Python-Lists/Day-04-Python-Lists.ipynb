🚀 Python Lists 
1. What is a List?

A list is a collection of multiple values stored in a single variable.

fruits = ["Apple", "Banana", "Mango"]

Why use a list?

Instead of:

fruit1 = "Apple"
fruit2 = "Banana"
fruit3 = "Mango"

You can store everything together:

fruits = ["Apple", "Banana", "Mango"]
2. List Properties

✅ Ordered (keeps insertion order)

✅ Mutable (can be changed)

✅ Allows duplicates

✅ Can store different data types

data = ["Dhruv", 22, 85.5, True]
3. Indexing

Every element has a position.

fruits = ["Apple", "Banana", "Mango"]
Value	Index
Apple	0
Banana	1
Mango	2

Access elements:

print(fruits[0])

Output:

Apple
4. Negative Indexing

Access elements from the end.

fruits = ["Apple", "Banana", "Mango"]
Value	Negative Index
Mango	-1
Banana	-2
Apple	-3
print(fruits[-1])

Output:

Mango
5. Modifying Elements

Lists are mutable.

fruits = ["Apple", "Banana", "Mango"]

fruits[1] = "Orange"

print(fruits)

Output:

['Apple', 'Orange', 'Mango']
🛠 Important List Methods
append()

Adds element at the end.

numbers = [1, 2, 3]

numbers.append(4)

Result:

[1, 2, 3, 4]
insert()

Adds element at a specific index.

numbers = [1, 2, 3]

numbers.insert(1, 100)

Result:

[1, 100, 2, 3]
remove()

Removes by value.

numbers = [10, 20, 30]

numbers.remove(20)

Result:

[10, 30]
pop()

Removes by index.

numbers = [10, 20, 30]

numbers.pop()

Result:

[10, 20]

Also returns the removed value:

removed = numbers.pop()
sort()

Sorts ascending.

numbers = [5, 2, 8, 1]

numbers.sort()

Result:

[1, 2, 5, 8]
reverse()

Reverses the list.

numbers = [1, 2, 3]

numbers.reverse()

Result:

[3, 2, 1]
🔁 Looping Through a List
marks = [80, 75, 90, 85]

for mark in marks:
    print(mark)

Flow:

mark = 80
mark = 75
mark = 90
mark = 85
📊 Functions Often Used with Lists
numbers = [10, 20, 30, 40]
Sum
sum(numbers)

Output:

100
Length
len(numbers)

Output:

4
Maximum
max(numbers)

Output:

40
Minimum
min(numbers)

Output:

10
Average
sum(numbers) / len(numbers)

Output:

25

```python
# Q23

# Count even and odd numbers.

numbers = [1,2,3,4,5,6,7,8]

# Expected:

# Even = 4
# Odd = 4
count_even = 0
count_odd = 0 
for i in numbers:
    if i%2 == 0:
        count_even = count_even + 1
    else:
        count_odd = count_odd + 1
print("Even : " , count_even)
print("Odd : " , count_odd)
```

```python
# Challenge 3

# Find how many times each number appears.

# Input:

numbers = [1,1,2,3,3,3]

# Output:

# 1 → 2 times
# 2 → 1 time
# 3 → 3 times
for i in numbers:
    print(f"{i} -> {numbers.count(i)} times")
    if numbers.count(i) > 1:
        for k in range(1,numbers.count(i)):
            numbers.remove(i)
```
