1️⃣ What is a Constructor?

A constructor is a special method that automatically executes when an object is created.

class Student:
    def __init__(self, name):
        self.name = name

s1 = Student("Dhruv")

print(s1.name)

Output:

Dhruv
2️⃣ Understanding self

self refers to the current object.

class Student:
    def __init__(self, name):
        self.name = name

s1 = Student("Dhruv")
s2 = Student("Rahul")

print(s1.name)
print(s2.name)

Output:

Rahul

```python
class Laptop:
    def __init__(self , brand , ram , price ) :
        self.brand = brand
        self.ram = ram
        self.price = price

    def display(self):
        print(f"brand : {self.brand}")
        print(f"RAM : {self.ram} GB")
        print(f"Price : {self.price}")

l1 = Laptop("Dell",8,50000)
l2 = Laptop("HP",8,51000)
l3 = Laptop("lenovo",32,65000)

l1.display()
l2.display()
l3.display()
```

```python
# Create an Employee class with name, salary, and department.
# Add a method display_info() to print employee details.
class Employee:
    def __init__(self , name , salary , department):
        self.name  = name
        self.salary = salary 
        self.department = department
    def display_info(self):
        print(f"Employee Name : {self.name}")
        print(f"Employee Salary : {self.salary}")
        print(f"Employee Department : {self.department}")
e1 = Employee("Dhruv" , 10000 , "data science")
e1.display_info()
```

```python
# Create a Laptop class and calculate the total value of three laptops.
class Laptop:
    def __init__(self , brand , ram , price ) :
        self.brand = brand
        self.ram = ram
        self.price = price

    def display(self):
        print(f"brand : {self.brand}")
        print(f"RAM : {self.ram} GB")
        print(f"Price : {self.price}")

l1 = Laptop("Dell",8,50000)
l2 = Laptop("HP",8,51000)
l3 = Laptop("lenovo",32,65000)

total = l1.price + l2.price + l3.price
print(total)
```

```python
# Create a Product class and apply a 10% discount using a method.
class Product:
    def __init__(self , product_name , price):
        self.name = product_name
        self.price = price
    def ten_discount(self):
        print(f"{(self.price * 10 )/100}")
p1 = Product("shampoo" , 2)
p1.ten_discount()
```

```python
# Create a Rectangle class and calculate area using object attributes.
class Rectangle:
    def __init__(self , leng , b ) : 
        self.length = leng 
        self.breadth = b
    def area(self):
        print(f"Area is : {self.length * self.breadth } ")
r1 = Rectangle(10 , 10 ) 
r1.area()
```

```python
# Create a Circle class and calculate circumference.
class Circle:
    def __init__(self , radius ) :
        self.radius = radius 
    def cir(self):
        print(f"Circumference is {2 * 3.14 * self.radius}")
c1 = Circle(10)
c1.cir()
```
