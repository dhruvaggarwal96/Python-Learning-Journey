```python
🚀 Day 12 — Polymorphism in Python
🎯 Today's Goal

By the end of today, you should be able to:

✅ Understand Polymorphism

✅ Use the same method name in different classes

✅ Understand Method Overriding

✅ Write flexible OOP code

✅ Build a small OOP project

1️⃣ What is Polymorphism?

The word Polymorphism means:

"Many Forms"

In Python, the same method can behave differently depending on the object.

Example:

class Dog:
    def sound(self):
        print("Bark")

class Cat:
    def sound(self):
        print("Meow")

Both classes have a method called sound().

But they produce different outputs.

2️⃣ Basic Example
class Dog:

    def sound(self):
        print("Bark")


class Cat:

    def sound(self):
        print("Meow")


d = Dog()
c = Cat()

d.sound()
c.sound()
Output
Bark
Meow

Same method name.

Different behavior.

This is Polymorphism.

3️⃣ Polymorphism with Loop

Instead of calling each object separately:

class Dog:
    def sound(self):
        print("Bark")

class Cat:
    def sound(self):
        print("Meow")

class Cow:
    def sound(self):
        print("Moo")
animals = [Dog(), Cat(), Cow()]

for animal in animals:
    animal.sound()
Output
Bark
Meow
Moo

The same code works for all objects.

4️⃣ Polymorphism Through Method Overriding
class Animal:

    def sound(self):
        print("Animal Sound")


class Dog(Animal):

    def sound(self):
        print("Bark")
d = Dog()
d.sound()
Output
Bark

The child class overrides the parent method.

5️⃣ Real-Life Example

Think of a Payment App.

Every payment method has:

pay()

But each works differently.

class UPI:

    def pay(self):
        print("Payment via UPI")


class CreditCard:

    def pay(self):
        print("Payment via Credit Card")


class Wallet:

    def pay(self):
        print("Payment via Wallet")
payments = [UPI(), CreditCard(), Wallet()]

for p in payments:
    p.pay()
💻 Practice Exercise 1

Create:

class Bird
class Parrot
class Eagle

Method:

fly()

Output:

Parrot flies low
Eagle flies high
💻 Practice Exercise 2

Create:

class Shape
class Circle
class Rectangle

Method:

area()

Each should print a different message.

🚀 Mini Project
Employee Management System
Requirements

Create:

Employee
Manager
Developer
Designer

Each class should have:

work()

Example:

Manager manages the team
Developer writes code
Designer creates UI designs

Store all objects inside a list.

Use a loop:

employees = [manager, developer, designer]

for emp in employees:
    emp.work()

This demonstrates polymorphism.

🎤 Interview Questions
Q1. What is Polymorphism?

Polymorphism means one interface (method name) having multiple forms or behaviors.

Q2. Why is Polymorphism useful?

It makes code:

Flexible
Reusable
Easy to maintain
Q3. Difference Between Inheritance and Polymorphism?
Inheritance	Polymorphism
Acquires properties from another class	Same method behaves differently
Focus on code reuse	Focus on flexibility
Q4. What is Method Overriding?

When a child class provides its own implementation of a parent class method.

Q5. Give a Real-Life Example of Polymorphism.

Payment systems:

UPI → pay()
Credit Card → pay()
Wallet → pay()

Same method name, different behavior.
```
